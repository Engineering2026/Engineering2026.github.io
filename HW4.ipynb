```python
from itertools import combinations
import math
points = [(2,3), (5,7), (1,8), (2,3), (9,1), (5,7)]
unique_points = list(set(points))
print(unique_points)

max_dist = 0
result = None

for p1, p2 in combinations(unique_points, 2):
    dist = math.dist(p1,p2)
    if dist > max_dist:
        max_dist = dist
        result = (p1,p2,dist)
print(result)
```

```python
import string

text ="The University of Missouri–St. Louis is a public research university rooted in the heart of the St. Louis region. As the largest public research institution in eastern Missouri, UMSL blends academic rigor with a strong commitment to community impact. Its mission centers on expanding access to high‑quality education, supporting economic development, and empowering students to build meaningful careers."

print(string.punctuation)
#remove punctuations
for p in string.punctuation:
  text=text.replace(p,"")

#convert to lower case
text=text.lower()
#convert from a string to a list of words
words = text.split()
print(words)

freq={}
for word in words:
    if word in freq:
      freq[word]+=1
    else:
      freq[word]=1
print(freq)

#method 2 for dictionary
#for w in words:
#  freq[w]=freq.get(w,0)+1
#print(freq)
#items() converts the dictionary into pairs (tuple of pairs)
#sort
freq_sorted=sorted(freq.items(),key=lambda x:x[1],reverse=True)
print(freq_sorted)

top5 = freq_sorted[:5]
print("top 5 ->",top5)
```

```python
students = {
"Alice": [88, 92, 79],
"Bob": [90, 85, 87],
"Charlie": [70, 80, 75]
}

def letter_grade(score):
  if score >=90:
    return "A"
  elif score >=80:
    return "B"
  elif score >=70:
    return "C"
  elif score >=60:
    return "D"
  else:
    return "F"
result = {}
for student, scores in students.items():
  average_score = sum(scores) / len(scores)
  result[student] = (average_score, letter_grade(average_score))

print(result)
#print(result.items())
sorted_students = sorted(result.items(), key=lambda x: x[1] [0], reverse=True)

print("Top Student is ", sorted_students[0])
```

```python
inventory = {
"laptop": {"price": 1200, "quantity": 4},
"mouse": {"price": 25, "quantity": 20},
"keyboard": {"price": 75, "quantity": 10}
}

total_value = 0
values = []

print(inventory.items())

for item, details in inventory.items():
  unit_price = details["price"]
  quantity = details["quantity"]
  value = unit_price * quantity
  values.append((item,value))
  total_value += value
print(values)
print("Total value of this inventory is",total_value)

#output most valuable product
most_valuable = max(values, key=lambda x: x[1])
print("Most valuable item is",most_valuable)

#sort it by values
sorted_products = sorted(values, key=lambda x: x[1], reverse=True)
print(sorted_products)

```
